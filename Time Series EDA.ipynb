```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla=pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head()
```

```python
df_tesla.tail()
```

```python
## xlimit and y limit
df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4))
```

```python
index=df_tesla.loc['2020-01-01':'2021-09-01'].index
share_open=df_tesla.loc['2020-01-01':'2021-09-01']['Open']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure,axis=plt.subplots()
plt.tight_layout()
## Preventing overlapping
figure.autofmt_xdate()
axis.plot(index,share_open)
```

```python
df_tesla=df_tesla.reset_index()
```

```python
df_tesla.info()
```

```python
df_tesla=df_tesla.set_index('Date',drop=True)
```

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min()['Open'].plot()
```

```python
df_tesla.resample(rule='QS').min()['Open'].plot()
```

```python
df_tesla.resample(rule='BA').min()['Open'].plot()
```

```python
# ##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
```

```python
df_tesla.resample(rule='BQA').min()['Open'].plot()
```

```python
##plotting
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

```python
df_tesla['High'].rolling(11).mean().head(20)
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla
```

```python
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,5))
```

```python
df_tesla['EWMA'] = df_tesla['Open'].ewm(alpha=0.1 , adjust = False).mean()
```

```python
df_tesla[['Open' , 'EWMA']].plot(xlim = ['2020-01-01' , '2021-01-01'] , figsize = (15,6))
```

```python
df_tesla['EWMA+0.3'] = df_tesla['Open'].ewm(alpha=0.3 , adjust = False).mean()
```

```python
df_tesla[['Open' ,'EWMA' ,  'EWMA+0.3']].plot(xlim = ['2020-01-01' , '2021-01-01'] , figsize = (15,6))
```
